```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
```

```python
#Opening the IMDb site 
driver.get("https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1")
```

```python
Sno=[]
Movie_Name=[]
Director_1=[]
Director_2=[]
Duration=[]
Year=[]
Ratings=[]
Metascore=[]
url=[]
```

```python
Star1=[]
Star2=[]
Star3=[]
Star4=[]
Genre1=[]
Genre2=[]
Genre3=[]
Votes=[]
Gross_Collection=[]
Popularity=[]
Certification=[]
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
start=0
end=35
for page in range(start,end):
    no=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
    for i in no:
        Sno.append(i.text) 
        time.sleep(1)
        
    vote=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in vote:
        try:
            Votes.append(i.text.split('|')[0].split(':')[1])
        except NoSuchElementException:
            Votes.append('-')
        time.sleep(1)
    nxt=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    nxt.click()
    time.sleep(1)
```

```python
len(Sno)
```

```python
start=0
end=35
for page in range(start,end):
    url_1=driver.find_elements(By.XPATH,'//h3/a')
    for i in url_1:
        a=i.get_attribute('href')
        url.append(a) 
    nxt=driver.find_element(By.XPATH,'//a[@class="lister-page-next next-page"]')
    nxt.click()
```

```python
for i in url:
    driver.get(i)

    
# Extracting Movie Name    
    try:
        name = driver.find_element(By.XPATH,'//span[@class="sc-afe43def-1 fDTGTb"]')                           
        Movie_Name.append(name.text)

    except NoSuchElementException:
        Movie_Name.append('-')
        
    
# Extracting Duration 
    try:
        time = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[3]')                           
        Duration.append(time.text)

    except NoSuchElementException:
        Duration.append('-')
        

# Extracting Year
    try:
        year = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a')                           
        Year.append(year.text)

    except NoSuchElementException:
        Year.append('-')

        
# Extracting Ratings
    try:
        Rating = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/div[1]/div/div[1]/a/span/div/div[2]/div[1]/span[1]')                           
        Ratings.append(Rating.text)

    except NoSuchElementException:
        Ratings.append('-')
        
# Extracting Metascore
    try:
        score = driver.find_element(By.XPATH,'//span[@class="score-meta"]')                           
        Metascore.append(score.text)

    except NoSuchElementException:
        Metascore.append('-')

        
# Extracting Director Name 1

    try:
        dir_1= driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li[1]/a')      
        Director_1.append(dir_1.text)

    except NoSuchElementException:
        Director_1.append('-')
        
# Extracting Director Name 2

    try:
        dir_2= driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li[2]/a')      
        Director_2.append(dir_2.text)

    except NoSuchElementException:
        Director_2.append('-')
    
# Extracting star1   
    try:
        star = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[1]/a')                           
        Star1.append(star.text)

    except NoSuchElementException:
        Star1.append('-')

# Extracting star2
    try:
        star2 = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[2]/a')                           
        Star2.append(star2.text)

    except NoSuchElementException:
        Star2.append('-')
        
# Extracting star3
    try:
        star3 = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[3]/a')                           
        Star3.append(star3.text)

    except NoSuchElementException:
        Star3.append('-')

# Extracting star4
    try:
        star4 = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div/ul/li[4]/a')                           
        Star4.append(star3.text)

    except NoSuchElementException:
        Star4.append('-')

# Extracting Genre1
    try:
        genre1 = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[1]')                           
        Genre1.append(genre1.text)

    except NoSuchElementException:
        Genre1.append('-')

# Extracting Genre2
    try:
        genre2 = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[2]')                           
        Genre2.append(genre2.text)

    except NoSuchElementException:
        Genre2.append('-')

# Extracting Genre3
    try:
        genre3 = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[1]/div[2]/a[3]')                           
        Genre3.append(genre3.text)

    except NoSuchElementException:
        Genre3.append('-')


# Extracting Popularity
    try:
        Popular = driver.find_element(By.XPATH,'//div[@class="sc-5f7fb5b4-1 bhuIgW"]')                           
        Popularity.append( Popular.text)

    except NoSuchElementException:
        Popularity.append('-')
        
# Extracting Certification
    try:
        certificate = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[2]/a')                           
        Certification.append(certificate.text)

    except NoSuchElementException:
        Certification.append('-')
        
# Extracting Gross Collection
    try:
        Collection = driver.find_element(By.XPATH,'/html/body/div[2]/main/div/section[1]/div/section/div/div[1]/section[12]/div[2]/ul/li[2]/div')                           
        Gross_Collection.append(Collection.text)

    except NoSuchElementException:
        Gross_Collection.append('-')
```
